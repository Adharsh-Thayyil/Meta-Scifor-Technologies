```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

n = pd.read_csv('/content/ucl-finals.csv')
print(n.to_string())
```

```python
winner_counts = n['winner'].value_counts()
plt.figure(figsize = (15,7))
winner_counts.plot(kind='bar', color='skyblue')
plt.title('Champions League Winners (1955–2022)')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
country_counts = n['winner-country'].value_counts()
country_counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Champions League Titles by Country (1955–2022)')
plt.ylabel('')
plt.show()
```
